# CORRELATION MATRIX

```python
#1st I will map the categorical variables value into numeric.

cleanup_nums = {"Gender":     {"Male": 1, "Female": 0},
                "Married": {"Yes": 1, "No": 0},
               "Dependents": {"0":0,"1":1,"2":2,"3+":3},
               "Education": {"Graduate":1,"Not Graduate":0},
               "Self_Employed": {"Yes":1,"No":0},
               "Property_Area": {"Rural":0,"Semiurban":1,"Urban":2},
               "Loan_Status": {"Y":1,"N":0}}
```

```python
obj_df = data.select_dtypes(include=['object']).copy()
obj_df.head()
```

```python
obj_df = obj_df.replace(cleanup_nums)
obj_df.head()
```

```python
num_df = data.select_dtypes(include=['int64','float64'])
num_df.head()
```

```python
df = pd.concat([obj_df,num_df],axis=1)
```

# OUTLIER DETECTION AND REMOVAL

```python
# using z score I will find the outliers
from scipy.stats import zscore
z = np.abs(zscore(df))
```

```python
threshold = 3
df_new =df[(z<threshold).all(axis = 1)]
df_new.shape
```

```python
df.shape
```

```python
data_1 = df_new.copy()
```

```python
#splitting data
X = data_1.drop('Loan_Status',axis=1)
X.head()
```

```python
Y = df_new['Loan_Status']
Y.head()
```

```python
#removing skewness of continues input variables.
X['ApplicantIncome'] = np.log1p(X['ApplicantIncome'])
X['CoapplicantIncome'] = np.log1p(X['CoapplicantIncome'])
X['LoanAmount'] = np.log1p(X['LoanAmount'])
```

```python
X.skew()
```

```python
#all other variables are categorical variables so we will not change them to reduce the skewness.
#Now I will create dummy variables for all categorical variables.

cat_dumm = pd.get_dummies(X, columns=['Gender', 'Married','Dependents','Education','Self_Employed',
                          'Credit_History'],drop_first=True)
cat_dumm.head()
```

```python
#converting loan amount term into years

cat_dumm['Loan_Amount_Term'] = cat_dumm['Loan_Amount_Term']/12
```

```python
#Model training, as this is a binary classification , I will try cross val score with different models

from sklearn.linear_model import LogisticRegression
from sklearn.ensemble import RandomForestClassifier, AdaBoostClassifier , GradientBoostingClassifier
from sklearn.neighbors import KNeighborsClassifier
from sklearn.svm import SVC
from sklearn.naive_bayes import GaussianNB
from sklearn.tree import DecisionTreeClassifier
from sklearn.metrics import accuracy_score,confusion_matrix,classification_report,roc_auc_score,roc_curve
from sklearn.model_selection import cross_val_score,GridSearchCV,train_test_split
```

```python
models = [LogisticRegression(),SVC(),DecisionTreeClassifier(),KNeighborsClassifier(),RandomForestClassifier(),
         GaussianNB(),AdaBoostClassifier(),GradientBoostingClassifier(),]

for model in models:
    print("model name =",model)
    model.scores = cross_val_score(model, cat_dumm, Y, cv=10)
    print('Model scores :',model.scores)
    print('Model mean score: ',model.scores.mean())
    print('Model std score :', model.scores.std())
    print('\n')
    print('********************************************')
```

we see that Logistic regression and GaussianNB giving us the best mean score, I will create my model seperately with Logistic regression and GuassianNB.

```python
LogisticRegression()
```

```python
lr_param = {'C':[0.1,0.01,0.001,1.0],'penalty':['l1', 'l2', 'elasticnet', 'none'],
        'random_state': range(41,100),
           'solver': ['newton-cg', 'lbfgs', 'liblinear', 'sag', 'saga']}
```

```python
grid_search = GridSearchCV(estimator = LogisticRegression(), param_grid = lr_param)
grid_search.fit(cat_dumm, Y)
grid_search.best_params_
```

```python
final_lr_model = LogisticRegression(C=0.1, penalty='l1', random_state=41, solver='liblinear')
```

```python
X_train,X_test,y_train,y_test = train_test_split(cat_dumm,Y,test_size=.20)
print('shape of X_train :', X_train.shape)
print('shape of X_test :', X_test.shape)
print('shape of y_train :', y_train.shape)
print('shape of y_test :', y_test.shape)
```

```python
final_model = final_lr_model.fit(X_train,y_train)
final_predict = final_lr_model.predict(X_test)
Accuracy = accuracy_score(y_test,final_predict)

print('Accuracy score :', Accuracy)

print('\n')

print(classification_report(y_test,final_predict))
```

```python
print('confusion metrix using logistic regression :', '\n', confusion_matrix(y_test,final_predict))
```

```python
#GuassianNB
gnb = GaussianNB()
gnb_model = gnb.fit(X_train,y_train)
gnb_predict = gnb.predict(X_test)
Accuracy_gnb = accuracy_score(y_test,gnb_predict)

print('Accuracy score :', Accuracy_gnb)

print('\n')

print(classification_report(y_test,gnb_predict))
```

```python
print('confusion metrix using logistic regression :', '\n', confusion_matrix(y_test,gnb_predict))
```

we are getting exact same results using Logistic regression and GaussianNB models.

```python
lr_pred_prob = final_lr_model.predict_proba(X_test)
lr_pred_prob
```

```python
#predicting the probability of having one
lr_pred_prob = final_lr_model.predict_proba(X_test)[:,1]
lr_pred_prob
```

```python
#fpr = false positive rate
#tpr = true positive rate
#thresholds we set

fpr,tpr,thresholds = roc_curve(y_test,lr_pred_prob)
auc = roc_auc_score(y_test,lr_pred_prob)
```
